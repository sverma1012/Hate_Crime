## Extract Race from all Years

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings('ignore')
```

```python
# 2009

race_09 = pd.read_excel('../2009/Table 1/Table1_v2_2009.xls', sheet_name=1)
race_09
```

```python
race_09.columns = ['bias', 'incidents', 'offenses', 'victims', 'offenders']
race_09.columns
```

```python
victims_09 = race_09[['bias', 'victims']]
victims_09
```

```python
victims_09['year'] = 2009
victims_09
```

```python
victims_09_v2 = pd.pivot_table(victims_09,
                               index=['year'],
                               columns=['bias'])

victims_09_v2
```
